## Plotting number of matches of each season

```python
match['year'] = match['date'].str[:4].astype(int) #getting year from date column
data = go.Histogram(x=match['year'], marker=dict(color = '#8be1c5', line=dict(width=0.5))) #making histogram with no of matches in year and giving it a color and border to each block
layout = go.Layout(title='No. of matches in every season', 
                   xaxis=dict(title='season',tickmode='linear'), # tickmode is used to display every year on x axiss
                   yaxis=dict(title='No. of matches'), bargap=0.4) #giving the layout to histogram
graph= go.Figure(data=data, layout=layout) #making the graph
iplot(graph) #plotting the graph
```

##### Analysis: The year 2013 has maximum number of matches. It can be because of super over. Also there were 10 teams in 2011, 9 in 2012 and 2013

## Total wins of a team

```python
# concat different columns of team 1 and team 2 to get match of each team
no_of_matches_played=pd.concat([match['team1'],match['team2']])
```

```python
no_of_matches_played
```

```python
# calculating the matches of each team
# When we reset the index, the old index is added as a column, and a new sequential index is used
no_of_matches_played=no_of_matches_played.value_counts().reset_index()
```

```python
no_of_matches_played
```

```python
# giving the names to coloumns
no_of_matches_played.columns=['Team','Total Matches']
```

```python
no_of_matches_played
```

```python
# calculating total wins of each team
wins = match['winner'].value_counts().reset_index()
```

```python
wins
```

```python
# giving names to wins table
wins.columns=['Team','winner']
```

```python
wins
```

```python
# merging the tables to get no of wins of each team
winners = pd.merge(no_of_matches_played,wins)
```

```python
winners
```

```python
# making table
table = go.Table(
    header=dict(values=["Team","Total Matches","Wins"], # giving the values of header of table
                fill = dict(color='#e9dd77')),
    cells=dict(values=[winners['Team'], winners['Total Matches'], winners['winner']],
              fill = dict(color='#e6e2c2'))) #providing the values to be filled in cells of table

layout = dict(
    title='Total Matches vs Wins per team',  
)

fig1 = dict(data=[table], layout=layout) #using list for data as it contains 2 values - header and cells 
iplot(fig1)
```

##### Analysis: It shows the total matches played by each team and no of matches won

## Percentage win of each team

```python
# making bar for team v/s total matches
trace1 = go.Bar(x=winners.Team,y=winners['Total Matches'],
                name='Total Matches',marker=dict(color='#d46d40'))

# making bar for team v/s no of wins
trace2 = go.Bar(x=winners.Team,y=winners['winner'],
                name='Matches Won',marker=dict(color='#1bafa6'))

# making bar for team v/s percentage win
trace3 = go.Bar(x=winners.Team,
               y=(round(winners['winner']/winners['Total Matches'],4)*100), #4 indiacates the number of digits for percentage
               name='Win Percentage',marker=dict(color='#af1b92'))

# combining the 3 values under one team
data = [trace1, trace2, trace3]

# providing layout to graph
layout = go.Layout(title='Match Played, Wins And Win Percentage',xaxis=dict(title='Team'),
                   yaxis=dict(title='Count'))

bar = go.Figure(data=data, layout=layout)
iplot(bar)
```

##### Analysis: It shows that CSK has highest win percentage followed by MI

## Does toss win matches?

```python
# adding a column toss_win_match in dataset
# if toss_winner = winner, then put yes else put no
match['toss_win_match'] = np.where((match.toss_winner 
                                    == match.winner),'Yes','No')
```

```python
# getting the count of yes and no
toss_match_win = match['toss_win_match'].value_counts()
```

```python
toss_match_win
```

```python
toss_win_match_labels = (np.array(toss_match_win.index)) #getting the 'yes' and 'no' in labels array
toss_win_match_values = (np.array((toss_match_win / toss_match_win.sum())*100)) # getting the percentage in array
```

```python
# data for toss_win_match pie chart
toss_win_match_data = go.Pie(labels = toss_win_match_labels,
                           values = toss_win_match_values)
# layout for toss pie chart
toss_win_match_layout = go.Layout(title='Does winning toss infers winning match?')

toss_win_match_pie = go.Figure(data = toss_win_match_data, layout = toss_win_match_layout)
iplot(toss_win_match_pie)
```

##### Analysis: Though winning toss gives you an advantage but it doesn't significantly implies that winning the toss helps in winning the game.

## Highest score of IPL

```python
# calulating the totoal score by grouping id, innings,batting_team, bowling_team
# if all four matches then calculate sum of total_runs
high_scores=ball.groupby(['id', 'inning','batting_team','bowling_team'])['total_runs'].sum().reset_index()
```

```python
high_scores
```

```python
# get scores of greater than 200
high_scores=high_scores[high_scores['total_runs']>=200]
```

```python
high_scores
```

```python
# getting the first 10 highest score
hss = high_scores.nlargest(10,'total_runs')
```

```python
hss
```

```python
# getting the data for table
hss_data = go.Table(
    header=dict(values=["Inning","Batting Team","Bowling Team", "Total Runs"], # values for head
                fill = dict(color = 'lightgrey')),
    cells=dict(values=[hss['inning'], hss['batting_team'], hss['bowling_team'], hss['total_runs']])) # values for cells

hss_layout = dict(
    title='Highest scores of IPL',   
)

hss_table = dict(data=[hss_data], layout=layout)
iplot(hss_table)
```

## Total and average runs per season

```python
# we took id and year column from match and merged it with ball dataset through id column
batsmen = match[['id','year']].merge(ball)
```

```python
batsmen
```

```python
# we calculated the sum of total runs grouped by year
season=batsmen.groupby(['year'])['total_runs'].sum().reset_index()
```

```python
season
```

```python
# we counted the number of matches through id and grouped it according to year
avgruns_each_season=match.groupby(['year']).count().id.reset_index()
```

```python
avgruns_each_season
```

```python
# renamed the column
avgruns_each_season.rename(columns={'id':'matches'},inplace=1)
```

```python
avgruns_each_season
```

```python
# added a column total_runs 
avgruns_each_season['total_runs']=season['total_runs']
```

```python
avgruns_each_season
```

```python
# added a coloumn for avg runs per per match of each season
avgruns_each_season['average_runs_per_match']=avgruns_each_season['total_runs']/avgruns_each_season['matches']
```

```python
avgruns_each_season
```

##### Analysis: maximum runs scored was in year 2013. one reason might be because there were 9 teams. There were high scoring match in year 2018 as its average runs per match is maximum
